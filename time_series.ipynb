```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import os
```

```python
file_feature = {"csv/NASA_GISS_Surface_Temperature_Analysis_Long_Term_Mean_37.1592_20.9047.csv" : 'air',
         "csv/NOAA_Extended_Reconstructed_SST_V5_Long_Term_Mean_37.1592_20.9047.csv" : 'sst',
         "csv/Global_Precipitation_Climatology_Project_(GPCP)_Monthly_Analysis_Product_37.1592_20.9047.csv" : 'precip',
         "csv/CHIRPS_V3_Monthly_Mean_37.1592_20.9047.csv" : 'precip',
         "csv/NOAA_Global_Surface_Temperature_Long_Term_Mean_37.1592_20.9047.csv" : 'air',
         "csv/NOAA_OI_SST_V2_High_Resolution_Dataset_37.1592_20.9047.csv" : 'sst',
         "csv/IAP_Salinity_Dataset_37.5_21.0.csv" : 'salinity'}
```

```python
def plot_feature(file, feature):
    df = pd.read_csv(file)
    plt.figure(figsize=(12, 6))
    if len(df["time"]) < 13:
        plt.plot(df["time"].str[5:7], df[feature], label=feature)
    else:
        plt.plot(df["time"].astype(str).str[2:4], df[feature], label=feature)
        plt.xticks(rotation=90)
    plt.title(f'{feature} at {file}')
    plt.xlabel('Date')
    plt.ylabel(feature)
    plt.legend()

    plt.savefig(f'plots/{feature}_{os.path.basename(file).replace(".csv", ".png")}', dpi=300)
```

```python
for file, feature in file_feature.items():
    plot_feature(file, feature)
```
