```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# Load GEBCO dataset
gebco = xr.open_dataset("gebco_2024_n30.0_s10.0_w30.0_e45.0.nc") 
print(gebco.data_vars)
```

```python
topo = gebco['elevation']
```

```python
fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection=ccrs.PlateCarree())

topo.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap='terrain',
    cbar_kwargs={'label': 'Elevation (m)'}
)

ax.set_extent([30, 45, 10, 30], crs=ccrs.PlateCarree())
ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS)
ax.set_title("Topography of the Red Sea Region")
ax.set_subtitle("GEBCO 2024 Dataset")

plt.show()

# Save the figure
fig.savefig("plots/gebco_red_sea_topography.png", dpi=300, bbox_inches='tight')
```
